```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",)
```

```python
# Function to generate a response from the model
def chat_with_model(prompt, max_new_tokens=256):
    # Add a system message to establish the role of the chatbot
    messages = [
        {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
        {"role": "user", "content": prompt},
    ]

    # Generate response using the pipeline
    outputs = pipeline(
        messages,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=0.6,
        top_p=0.9,
    )

    # Extract the generated text from the output
    response_text = outputs[0]["generated_text"]
    return response_text

def main():
    print("Welcome to the Meta-Llama-3-8B-Instruct chat! Type 'exit' to end the conversation.")
    user_input = input("You: ")

    while user_input.lower() != "exit":
        model_response = chat_with_model(user_input)
        print(f"Model: {model_response}")
        user_input = input("You: ")

if __name__ == "__main__":
    main()
```
